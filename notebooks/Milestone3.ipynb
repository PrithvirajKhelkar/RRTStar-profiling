```python
import matplotlib.pyplot as plt
```

```python
all_functions = ["Pop_Heap_Hole", "NearestNeighbor", "isStateValid", "Distance", "isCostBetterThan", "less_operation"]
all_inputs = ["Map", "3", "6", "8"]
data_plot1 = ["Total CPU Time", "Total CPU-Self Time"]
data_plot2 = ["Instructions Retired", "Number of Cycles"]
```

```python
mapping = {
    "3": "a",
    "6": "b",
    "8": "c",
    "Map": "d"
}
```

```python
input = "Map"

intructions_retired = []
for function in all_functions:
    intructions_retired.append(float(functions[function][input]["Total CPU Time"]))

number_of_cycles = []
for function in all_functions:
    number_of_cycles.append(float(functions[function][input]["Total CPU-Self Time"]) )
```

```python
print(intructions_retired)
print(number_of_cycles)
```

```python
import numpy as np

data_points = {
    'Total CPU Time': intructions_retired,
    'Total CPU-Self Time': number_of_cycles
}

x = np.arange(len(all_functions))  # the label locations
width = 0.33  # the width of the bars
multiplier = .5

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in data_points.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3, rotation=90)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Milliseconds (ms)')
ax.set_title(f'Input: {mapping[input]}\nRuntime Measurement per Subfunction')
ax.set_xticks(x + width, all_functions, rotation=30)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 380)

plt.savefig(f'figures/{input}_measurement_data.png')
plt.show()
```
